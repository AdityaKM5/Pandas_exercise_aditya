```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Simple training data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
```

```python
def compute_cost(w, b, x, y):
    m = len(x)
    total_cost = 0
    
    for i in range(m):
        prediction = w * x[i] + b
        total_cost += (prediction - y[i])**2
        
    return total_cost / (2 * m)
```

```python
w_vals = np.linspace(-1, 5, 100)
b_vals = np.linspace(-1, 10, 100)

W, B = np.meshgrid(w_vals, b_vals)
J = np.zeros(W.shape)
```

```python
for i in range(len(w_vals)):
    for j in range(len(b_vals)):
        J[j, i] = compute_cost(W[j, i], B[j, i], x, y)
```

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(W, B, J, cmap='viridis')

ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Cost J(w,b)')
ax.set_title('Cost Function Surface')

plt.show()
```

```python
plt.figure(figsize=(8,6))

contour = plt.contour(W, B, J, levels=30, cmap='viridis')
plt.clabel(contour, inline=True, fontsize=8)

plt.xlabel('Weight (w)')
plt.ylabel('Bias (b)')
plt.title('Contour Plot of Cost Function')

plt.show()
```
